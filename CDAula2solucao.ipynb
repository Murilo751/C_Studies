### Medio

```python
# Declarar variáveis
nome_analista = "Joana"
setor = "Inteligência de Mercado"
relatorios_entregues = 4

print(f"Olá, {nome_analista}! Você está alocada no setor de {setor} e já entregou {relatorios_entregues} relatórios.")

# Classificação de desempenho com estruturas condicionais
if relatorios_entregues <= 2:
    print("Em treinamento")
elif relatorios_entregues <= 5:
    print("Desempenho regular")
else:
    print("Desempenho excelente")

# Lista de datasets e impressão com for + range
datasets = ["vendas_janeiro.csv", "clientes.csv", "produtos.csv", "estoque.csv", "financeiro.csv"]

for i in range(len(datasets)):
    print(f"Dataset {i + 1}: {datasets[i]}")

# Simulação de conexão com API (sucesso na 3ª tentativa)
tentativas = 0
sucesso = False

while not sucesso:
    tentativas += 1
    print(f"Tentativa {tentativas} de conexão com a API...")

    if tentativas == 3:
        sucesso = True
        print("Conexão bem-sucedida!\n")

# Classe Analista com métodos apresentar() e adicionar_projeto()
class Analista:
    def __init__(self, nome, setor, projetos):
        self.nome = nome
        self.setor = setor
        self.projetos = projetos

    def apresentar(self):
        print(f"Olá, meu nome é {self.nome} e trabalho no setor de {self.setor}.")

    def adicionar_projeto(self, projeto):
        self.projetos.append(projeto)
        print(f"Projeto '{projeto}' adicionado com sucesso!")

# Criando objeto e utilizando os métodos
a = Analista("Joana", "Dados", [])

a.apresentar()
a.adicionar_projeto("Análise de Vendas Q1")
a.adicionar_projeto("Relatório de Clientes Ativos")
```

###Dificil

```python
# Importações necessárias
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem
imagem = cv2.imread('amostra1.jpg')

# Converter de BGR para RGB
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

# Exibir a imagem
plt.imshow(imagem_rgb)
plt.title("Imagem carregada")
plt.axis("off")
plt.show()

# Obter altura e largura da imagem
altura = imagem_rgb.shape[0]
largura = imagem_rgb.shape[1]

print(f"A imagem tem {altura} pixels de altura e {largura} de largura")

# Verificar se a imagem é RGB ou escala de cinza
if len(imagem_rgb.shape) == 3 and imagem_rgb.shape[2] == 3:
    print("Imagem em RGB")
else:
    print("Imagem em escala de cinza")

# Calcular a média do canal vermelho (usando amostragem de 10 em 10 pixels)
soma_vermelho = 0
contador = 0

for i in range(0, altura, 10):
    for j in range(0, largura, 10):
        pixel = imagem_rgb[i][j]
        vermelho = pixel[0]  # canal vermelho (RGB)

        soma_vermelho += vermelho
        contador += 1

media_vermelho = soma_vermelho / contador
print(f"Média do canal vermelho (amostragem): {media_vermelho:.2f}")

# Repetir tentativa até que altura seja maior que 300 pixels
tentativas = 0

while altura <= 300:
    print(f"Tentativa {tentativas + 1}: imagem com altura {altura}")
    tentativas += 1
    altura += 50  # simulação de redimensionamento

print("Altura adequada alcançada. Continuando processamento...")

# Classe AnalisadorDeImagem
class AnalisadorDeImagem:
    def __init__(self, nome_da_imagem, largura, altura):
        self.nome_da_imagem = nome_da_imagem
        self.largura = largura
        self.altura = altura

    def resumo(self):
        print(f"Imagem: {self.nome_da_imagem}")
        print(f"Largura: {self.largura}")
        print(f"Altura: {self.altura}")

    def avaliar_dimensao(self):
        if self.largura < 300:
            print("Imagem pequena")
        elif self.largura <= 800:
            print("Imagem média")
        else:
            print("Imagem grande")

# Criar e usar o objeto
img = AnalisadorDeImagem("amostra.jpg", largura, altura)
img.resumo()
img.avaliar_dimensao()
```
